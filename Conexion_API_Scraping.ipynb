```python
import requests

URL = "http://aca-web.gencat.cat/sdim2/apirest/data/EMBASSAMENT-EST"
response = requests.get(URL)

if response.status_code == 200:
    print('Solicitud exitosa')
    print('Data:', response.json())
else:
    print('Error en la solicitud, detalles:', response.text)
```

```python
import json
import pandas as pd

# Leer los datos del archivo JSON
data = response.json()

# Convertir los datos a un DataFrame
sensor_data = []

# Recorrer cada sensor y sus observaciones
for sensor in data['sensors']:
    sensor_id = sensor['sensor']
    for observation in sensor['observations']:
        sensor_data.append({
            'Sensor ID': sensor_id,
            'Value': observation['value'],
            'Timestamp': observation['timestamp'],
            'Location': observation['location']
        })

# Crear un DataFrame
df = pd.DataFrame(sensor_data)
df
```

```python
from bs4 import BeautifulSoup
```

```python
# URL de la página
url = 'https://www.lavanguardia.com/'

# Solicitud GET
response = requests.get(url)

# Verifica que la solicitud fue exitosa
if response.status_code == 200:
    # Analiza el contenido HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encuentra todos los títulos de los artículos
    titles = soup.find_all('h2', class_='title')

    # Imprime los títulos encontrados
    for title in titles:
        print(title.get_text(strip=True))
else:
    print(f'Error al acceder a la página: {response.status_code}')
```
