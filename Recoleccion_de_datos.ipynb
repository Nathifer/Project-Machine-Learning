**5. Consideraciones sobre los datos sensibles**

* Tipod e de datos sensibles:
  - Información Personal Identificable :  Estat civil (marital), Ocupació (job).
  - Información Financera Sensible: saldo medio anual(balance), Tiene prestamo hipotecario  (housing), tiene prestamo personal (loan).
  - Datos comportamentales: duracion de la comunicación (duration), Numero de contactos (campaign), dias desde el ultimo contacto (pdays).

* Medidas de Protección:
  - Anonimización y Pseudonimización:
  Para datos como el estado civil y la ocupación, se puede aplicar pseudonimización, sustituyendo los valores por códigos. En el caso del saldo medio anual, se pueden aplicar técnicas de anonimización, agregando datos o utilizando intervalos en lugar de valores exactos.

  - Acceso Restringido:
  El acceso a los datos sensibles estará limitado al personal autorizado, con una necesidad justificada podrá acceder a esta información.

  - Cumplimiento de Regulaciones:
  Cumplimiento del  Reglamento General de Protección de Datos.
